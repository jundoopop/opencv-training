```python
import cv2
import numpy as np
```

```python

def sobel_edge_detection(image):
  """Sobel edge detection.

  Args:
    image: The input image.

  Returns:
    The edge image.
  """

  # Convert the image to grayscale.
  grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  gx = np.array([[-1,0,1], [-2,0,2], [-1,0,1]])
  gy = np.array([[-1,-2,-1],[-2,0,2], [-1,0,1]])

  # Calculate the Sobel derivatives in the x and y directions.
  sobel_x = cv2.filter2D(grayscale_image, -1, gx)
  sobel_y = cv2.filter2D(grayscale_image, -1, gy)

  

  # Calculate the magnitude of the gradient.
  gradient_magnitude = np.sqrt(np.square(sobel_x) + np.square(sobel_y))

  # Normalize the gradient magnitude to the range [0, 255].
  gradient_magnitude = np.uint8(255 * gradient_magnitude / gradient_magnitude.max())

  return gradient_magnitude

```
